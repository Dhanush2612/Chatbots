```python
# function to create the chatbot
# we have the read_only to false so the chatbot learns from the user input as 
def create_bot(name):
    from chatterbot import ChatBot
    Bot = ChatBot(name = name,
                  read_only = False,                  
                  logic_adapters = ["chatterbot.logic.BestMatch"],                 
                  storage_adapter = "chatterbot.storage.SQLStorageAdapter")
    return Bot

# function to train the bot with a variety of topics
# the language we have chosen is english
# we can train the bot for other languages as well
def train_all_data(Bot):
    from chatterbot.trainers import ChatterBotCorpusTrainer
    corpus_trainer = ChatterBotCorpusTrainer(Bot)
    corpus_trainer.train("chatterbot.corpus.english")

# function to train the bot with custom data
# it uses ListTrainer to train data from lists 
def custom_train(Bot, conversation):
    from chatterbot.trainers import ListTrainer
    trainer = ListTrainer(Bot)
    trainer.train(conversation)

# function to start and take responses from the chatbot
# the chatbot stays running unless a word is typed from the bye_list 
def start_chatbot(Bot):
    print('\033c')
    print("Hello, I am Jarvis. How can I help you")
    bye_list = ["bye jarvis", "bye", "good bye" , "exit"] 
    
    while (True):
        user_input = input("me: ")   
        if user_input.lower() in bye_list:
            print("Jarvis: Good bye and have a blessed day!")
            break
        
        response = Bot.get_response(user_input)
        print("Jarvis:", response)
```

```python
# create chatbot 
home_bot = create_bot('Jarvis')

# check identity
identity = input("State your identity please: ")

# rules for responding to different identities
if identity == "Dhanush":
    print("Welcome, Dhanush. Happy to have you at home.")

elif identity == "Kai":
    print("Dhanush is out right now, but you are welcome to the house.")

else:
    print("Your access is denied here.")
    exit()

# custom data
house_owner = [
    "Who is the owner of this house?",
    "Dhanush is the owner of this house."
]
custom_train(home_bot, house_owner)

print("------ Training custom data ------")
# write and train your custom data here IF the identity is Dhanush
if identity == 'Dhanush':   

    city = [
        "Where do I live?",
        "Dhanush, you live in Bangalore."
    ]

    dob = [
        "What is my date of birth?",
        "Dhanush, you were born on 26th of December in the year 2000 ."
    ]
    fav_color = [
        "What is my favourite color?",
        "Dhanush, your favorite color is blue."
    ]

    fav_book = [
        "What is my favourite book?",
        "Dhanush, you prefer watching movies rather than reading books."
    ]

    fav_movie = [
        "What is my favourite sitcom?",
        "Dhanush, you don't have any specific favorite movie because you just love all movies!"
    ]

    fav_sitcom = [
        "What is my favourite sitcom?",
        "Dhanush, you love F.R.I.E.N.D.S"
    ]

    fav_sports = [
        "What is my favourite sport?",
        "Dhanush, you love 24 Hours of LeMans"
    ]
    # train chatbot with your custom data
    custom_train(home_bot, city)
    custom_train(home_bot, dob)
    custom_train(home_bot, fav_color)
    custom_train(home_bot, fav_book)
    custom_train(home_bot, fav_movie)
    custom_train(home_bot, fav_sitcom)
    custom_train(home_bot, fav_sports)

# start chatbot
start_chatbot(home_bot)
```
